# PROVA DE MACHINE LEARNING

```python
import pandas as pd

df = pd.read_csv('class01.csv')

df.head()
```

```python
X = df.drop(columns=['target'])
X.head()
```

```python
X.shape
```

```python
y = df['target']
y.head()
```

# NAIVE BAYES GAUSSIANO

```python
X_train = X[:350]
y_train = y[:350]

X_test = X[350:]
y_test = y[350:]
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python

model = GaussianNB()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
# acuracia do treino
y_true = y_train
y_pred = model.predict(X_train)

accuracy_score(y_true, y_pred)
```

```python
# utilizando o score

model.score(X_train, y_train)
```

```python
# acuracia pra validacao (teste)
model.score(X_test, y_test)
```

# Leave One Out

```python
from sklearn.model_selection import LeaveOneOut

my_cv = LeaveOneOut()
scores_train = []
scores_test = []


for train_index, test_index in my_cv.split(X):
    
    X_train = X.iloc[train_index]
    X_test = X.iloc[test_index]
    
    y_train = y.iloc[train_index]
    y_test = y.iloc[test_index]
    
    model = GaussianNB()
    model.fit(X_train, y_train)
    
    scores_train.append(model.score(X_train, y_train))
    scores_test.append(model.score(X_test, y_test))
    
```

```python
import numpy as np
```

```python
print(np.mean(scores_train))
print(np.mean(scores_test))
```

```python
from sklearn.model_selection import cross_val_score, cross_validate
```

```python
np.mean(cross_val_score(GaussianNB(), X, y, cv=LeaveOneOut()))
```

```python
results = cross_validate(GaussianNB(), X, y, cv=LeaveOneOut(), return_train_score=True)
print(np.mean(results['train_score']))
print(np.mean(results['test_score']))
```

# Exercicio 3

```python
df = pd.read_csv('class02.csv')
X = df.drop(columns=['target'])
y = df['target']
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import KFold
```

```python
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler().fit(X)
```

```python
X_std = scalar.transform(X)
```

```python
X.describe()
```

```python
pd.DataFrame(X_std)
```

```python
X_std = scalar.transform(X)

results = cross_validate(KNeighborsClassifier(n_neighbors=10), 
                         X_std, 
                         y, 
                         cv=KFold(n_splits=10), 
                         return_train_score=True)

print(np.mean(results['test_score']))
```

# Regressao

```python
df = pd.read_csv('reg01.csv')
X = df.drop(columns=['target'])
y = df['target']
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso()
scalar = StandardScaler().fit(X)
```

```python


results = cross_validate(lasso, 
                         scalar.transform(X), 
                         y, 
                         cv=LeaveOneOut(), 
                         return_train_score=True,
                         scoring='neg_root_mean_squared_error')

print(np.mean(results['train_score']))
print(np.mean(results['test_score']))
```
